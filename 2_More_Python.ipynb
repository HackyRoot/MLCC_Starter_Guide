## Confusion Matrix

A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known. 

![Confusion matrix](https://www.dataschool.io/content/images/2015/01/confusion_matrix_simple2.png)

Let's take an example of diabetes disease dataset, above table represents the confusion matrix for the same dataset.

### What can we learn from this matrix?

*   There are two possible predicted classes: "yes" and "no". If we were predicting the presence of a disease, for example, "yes" would mean they have the disease, and "no" would mean they don't have the disease.

*   The classifier made a total of 165 predictions (e.g., 165 patients were being tested for the presence of that disease).

*  Out of those 165 cases, the classifier predicted "yes" 110 times, and "no" 55 times. In reality, 105 patients in the sample have the disease, and 60 patients do not.


### Let's now define the most basic terms, which are whole numbers (not rates):


*    **true positives (TP)** : These are cases in which we predicted yes (they have the disease), and they do have the disease.
*    **true negatives (TN)** : We predicted no, and they don't have the disease.
*    **false positives (FP)** : We predicted yes, but they don't actually have the disease. (Also known as a "Type I error.")
*   ** false negatives (FN)**  : We predicted no, but they actually do have the disease. (Also known as a "Type II error.")

## Plotting

The most important function in matplotlib is plot, which allows you to plot 2D data. Here is a simple example:



```
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1) # divide the range of [0, 9.4] with the step size of 0.1 and generate a list out of it
y = np.sin(x) # sin fucnction provided by numpy

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
print(x)
```

With just a little bit of extra work we can easily plot multiple lines at once, and add a title, legend, and axis labels:

```
y_sin = np.sin(x) # sin function provided by numpy
y_cos = np.cos(x) # cos function provided by numpy

# Plot the points using matplotlib
plt.plot(x, y_sin) # plot x and y_sin using default line style and color
plt.plot(x, y_cos) # plot x and y_cos using default line style and color

plt.xlabel('x axis label') # x axis label
plt.ylabel('y axis label') # y axis label

plt.title('Sine and Cosine') # title of plot
plt.legend(['Sine', 'Cosine']) # places a legend on the axes.

plt.show()
```

## Colormap - cmap
 Colormap instances are used to convert data values (floats) from the interval [0, 1] to the RGBA color that the respective Colormap represents. 

```
n = 100 # don't worry much about this code for now
r = 2 * np.random.rand(n)
theta = 2 * np.pi * np.random.rand(n)
area = 200 * r**2 * np.random.rand(n)
colors = theta
```

You can see in code below that current cmap is = RdYlGn

Try to replace it with Blues, BrBG.

```
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.BrBG) # replace cmap with Blues, BrBG, RdYlGn
```

## Subplot and GridSpec

The Matplotlib **subplot()** function can be called to plot two or more plots in one figure.

```
for i in range(1, 7):
    plt.subplot(2, 3, i) # 2 rows, 3 columns, i index
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

The Matplotlib **subplot() **  is used when you want to have subplots of different sizes.


```
fig = plt.figure(2)

gs = gspec.GridSpec(nrows=3, ncols=3)
gs.update(left=0.05, right=0.48, hspace=0.05) # spacing

ax1 = plt.subplot(gs[0:2, 0:3]) # spans over 2 rows, 3 columns
ax1.text(x=0.5, y=0.5, s='ax1', va='center', ha='center')

ax2 = plt.subplot(gs[2, 0:2]) # spans 2 columns
ax2.text(x=0.5, y=0.5, s='ax2', va='center', ha='center')

ax3 = plt.subplot(gs[2, 2]) # default values
ax3.text(x=0.5, y=0.5, s='ax3', va='center', ha='center')

plt.suptitle('Subplot via multiple GridSpec with spans')
plt.subplots_adjust(top=0.90)
plt.show()

```
